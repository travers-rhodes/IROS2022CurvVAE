```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
```

```python
def dToR(degrees):
    return degrees/180. * np.pi
```

```python
theta = np.linspace(35,55,10)
z = np.linspace(-1,1,10)

thv, zv = np.meshgrid(theta,z)
```

```python
def compute_xyz(thetas, zs):
    rv = 10 +(zs+1)*(zs-1)*np.sin(dToR(thetas * 15+60))

    xv = rv * np.cos(dToR(thetas))
    zv = rv * np.sin(dToR(thetas))
    yv = zs
    return(xv,yv,zv)
```

```python
xv,yv,zv = compute_xyz(thv,zv)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter(xv,yv,zv)
#ax.set_xlim(-8,-12)
#ax.set_zlim(-2,2)
#ax.set_ylim(-2,2)
```
